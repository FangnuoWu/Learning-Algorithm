```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure()
figsize=(100, 100)
plt.xlabel("size of data(n)")
plt.ylabel("false positive propobility(p)")
plt.title("analysize the false positive rates of two bloom filter")
data = np.loadtxt("data.txt")
plt.plot(data[:,0],data[:,1]/data[:,0],linewidth=0.1,color="r",label='first type')
plt.plot(data[:,0],data[:,2]/data[:,0],linewidth=0.1,color="b",label='second type')
plt.xlim(xmin=0)
plt.ylim((0, 1))
plt.show()
```

```python
import matplotlib.ticker as mticker  
plt.figure()
plt.xlabel("size of data(n)")
plt.ylabel("false positive propobility(p)")
plt.title("analysize the false positive rates of two bloom filter")

data = np.loadtxt("100M.txt")
pow = 2**20
L1,= plt.plot(data[:,0]/pow,data[:,1]/data[:,0],linewidth=1,color="r",label='first type')
L2,= plt.plot(data[:,0]/pow,data[:,2]/data[:,0],linewidth=1,color="b",label='second type',linestyle='--', dashes=(10, 10))

plt.xlim(xmin=0)
plt.ylim((0, 1))
plt.gca().xaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f M'))
plt.legend([L1,L2],['type one','type two'],loc='lower right')

plt.savefig("10M.png",dpi=400)
plt.show()
```
